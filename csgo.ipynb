```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("csgo.csv")
data.head(5)
```

```python
x = data.drop(columns = {'result', 'year', 'month', 'day', 'team_a_rounds', 'team_b_rounds', 'date', 'wait_time_s', 'match_time_s'})
y = data['result']
```

```python
print(type(x))
```

```python
y.head(5)
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
map_encoded = encoder.fit_transform(x[['map']])

map_encoded_df = pd.DataFrame(map_encoded, 
                              columns=encoder.get_feature_names_out(['map']),
                              index=x.index)
X = pd.concat([x.drop(columns=['map']), map_encoded_df], axis=1)
```

```python
X.head(5)
print(type(X))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(y)
Y = Y.reshape(-1,1)
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 100, shuffle = False)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
clf = SVC()
clf.fit(X_train, Y_train)
```

```python
predict = clf.predict(X_test)
for i, j in zip(Y_test, predict):
    print("label : {} du doan : {}".format(i,j))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test, predict, labels=[0,1,2]))
```
